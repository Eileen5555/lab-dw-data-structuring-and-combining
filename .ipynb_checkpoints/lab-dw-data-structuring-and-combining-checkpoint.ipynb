```python
import pandas as pd
df_1=pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df_2=pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df_3=pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")

# colums df_1
df_1.columns = df_1.columns.str.replace(' ', '_').str.lower()
df_1.rename(columns={'st': 'state'}, inplace=True)
# colums df_2
df_2.columns = df_2.columns.str.replace(' ', '_').str.lower()
df_2.rename(columns={'st': 'state'}, inplace=True)
# colums df_3
df_3.columns = df_3.columns.str.replace(' ', '_').str.lower()

customer_data = pd.concat([df_1,df_2,df_3])

#cleaning: 
import pandas as pd 
customer_data = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')

def replace_exact_match(value, to_replace, replacement):
    if value == to_replace:
        return replacement
    return value

customer_data.columns = customer_data.columns.str.replace(' ', '_').str.lower()
customer_data.rename(columns={'st': 'state'}, inplace=True)
customer_data.rename(columns={'st': 'state'}, inplace=True)
# Column Gender:
customer_data['gender'] = customer_data['gender'].str.replace("Male","M")
customer_data['gender'] = customer_data['gender'].str.replace("Femal","F")
customer_data['gender'] = customer_data['gender'].str.replace("female","F")
# Column state: 
customer_data['state'] = customer_data['state'].str.replace("WA","Washington")
customer_data['state'] = customer_data['state'].str.replace("AZ","Arizona")
customer_data['state'] = customer_data['state'].apply(replace_exact_match, args=("Cali", "California"))
# Column education:
customer_data['education'] = customer_data['education'].apply(replace_exact_match, args=("Bachelors", "Bachelor"))
# Customer Lifetime
customer_data['customer_lifetime_value'] = customer_data['customer_lifetime_value'].str.replace("%","")
#vehicle class
customer_data['vehicle_class'] = customer_data['vehicle_class'].apply(replace_exact_match, args=("Luxury SUV", "Luxury"))
customer_data['vehicle_class'] = customer_data['vehicle_class'].apply(replace_exact_match, args=("Luxury Car", "Luxury"))

#Customer lifetime
customer_data['customer_lifetime_value'] = customer_data['customer_lifetime_value'].astype(float)

#Number of open complaints 
customer_data['number_of_open_complaints'] = customer_data['number_of_open_complaints'].apply(replace_exact_match, args=("1/0/00", 0))
customer_data['number_of_open_complaints'] = customer_data['number_of_open_complaints'].apply(replace_exact_match, args=("1/1/00", 1))
customer_data['number_of_open_complaints'] = customer_data['number_of_open_complaints'].apply(replace_exact_match, args=("1/2/00", 2))
customer_data['number_of_open_complaints'] = customer_data['number_of_open_complaints'].apply(replace_exact_match, args=("1/3/00", 3))
customer_data['number_of_open_complaints'] = customer_data['number_of_open_complaints'].apply(replace_exact_match, args=("1/4/00", 4))
customer_data['number_of_open_complaints'] = customer_data['number_of_open_complaints'].apply(replace_exact_match, args=("1/5/00", 5))

#Drop the rows or columns with null values
customer_data = customer_data.dropna()

display(customer_data)
print(customer_data.shape)

# data types & suggestions for fixing any incorrect data types
print(customer_data.dtypes)
print("\nCorrection Suggestions:\nCustomer: string\nST: string\nGender: string\nEducation: string\n Customer Lifetime Value: float (convert NaN into zero)\nNumber of Open Complaints: integer\n Policy Type: String\nVehicle Class: string")
# number of unique values for each column + determine which columns appear to be categorical
print("\nUnique Values:")
for column in customer_data.columns:
    unique_values = customer_data[column].nunique()
    print("Column:", column, "\nUnique values: ", unique_values,)
    if unique_values < 10:
        print(customer_data[column].unique())

# summary statistics such as mean, median, mode, standard deviation, and quartiles & conclusions
numerical = customer_data.describe()
print("\nDescriptive Statistics for Numerical Variables:")
print(numerical)
print("\n")
#summary statistics for categorical & conclusions
categorical = customer_data.select_dtypes(include=['object']).columns
for column in categorical:
    unique_values_count =  customer_data[column].nunique()
    if unique_values_count < 10:
        frequency_counts =  customer_data[column].value_counts()
        print("\nFrequency counts for " + column + ":") 
        print(frequency_counts)


```
